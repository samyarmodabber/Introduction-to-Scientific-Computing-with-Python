# Matplotlib

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(-5, 2, 100) # create our x
y1 = x**3 + 5*x**2 + 10
y2 = 3*x**2 + 10*x
y3 = 6*x + 10
fig, ax = plt.subplots()

ax.plot(x, y1, color="blue", label="y(x)")
ax.plot(x, y2, color="red", label="y'(x)")
ax.plot(x, y3, color="green", label="y‚Äù(x)")

ax.set_xlabel("x")
ax.set_ylabel("y")
ax.legend()
```

```python
fig = plt.figure(figsize=(8, 2.5), facecolor="#f1f1f1")
# axes coordinates as fractions of the canvas width and height
left, bottom, width, height = 0.1, 0.1, 0.8, 0.8
ax = fig.add_axes((left, bottom, width, height),facecolor="#e1e1e1")

x = np.linspace(-2, 2, 1000)
y1 = np.cos(40 * x)
y2 = np.exp(-x**2)

ax.plot(x, y1 * y2, color="blue")
ax.plot(x, y2, 'g')
ax.plot(x, -y2, 'g')

ax.set_xlabel("x")
ax.set_ylabel("y")

fig.savefig("graph.pdf", dpi=100, facecolor="#f1f1f1")
```

```python
x = np.linspace(-10, 10, 500)
one = np.ones(500)

fig, axes = plt.subplots(1, 4, figsize=(16,3))

axes[0].plot(x, one )
axes[1].plot(x, x)
axes[2].plot(x, x**2)
axes[3].plot(x, x**3)

```

```python
x = np.linspace(0, 50, 500)
y = np.sin(x) * np.exp(-x/10)
fig, ax = plt.subplots(figsize=(8,5), subplot_kw={'facecolor':"#ebf5ff"})

ax.plot(x, y, lw=1)

ax.set_xlabel("x", labelpad=5, fontsize=18, fontname='serif', color="blue")
ax.set_ylabel("f(x)", labelpad=15, fontsize=18, fontname='serif', color="brown")
ax.set_title("axis labels and title example", fontsize=16, fontname='serif', color="red")
```

```python
import matplotlib as mpl
fig, axes = plt.subplots(1, 3, figsize=(12, 4))
x = np.linspace(0, 50, 500)
y = np.sin(x) * np.exp(-x/10)

x_major_ticker = mpl.ticker.MultipleLocator(4)
x_minor_ticker = mpl.ticker.MultipleLocator(1)
y_major_ticker = mpl.ticker.MultipleLocator(0.25)
y_minor_ticker = mpl.ticker.MultipleLocator(0.05)
for ax in axes:
 ax.plot(x, y, lw=2)
 ax.xaxis.set_major_locator(x_major_ticker)
 ax.yaxis.set_major_locator(y_major_ticker)
 ax.xaxis.set_minor_locator(x_minor_ticker)
 ax.yaxis.set_minor_locator(y_minor_ticker)

axes[0].set_title("default grid")
axes[0].grid()

axes[1].set_title("major/minor grid")
axes[1].grid(color="blue", which="both", linestyle=':',linewidth=0.5)

axes[2].set_title("individual x/y major/minor grid")
axes[2].grid(color="red", which="major", axis='x', linestyle='-', linewidth=0.5)
axes[2].grid(color="red", which="minor", axis='x', linestyle=':', linewidth=0.25)
axes[2].grid(color="red", which="major", axis='y', linestyle='-',linewidth=0.5)
axes[2].grid(color="red", which="minor", axis='y', linestyle=':',linewidth=0.5)
```
